# Simulatory:
1. 100000000 her co vypadají jinak, ale fungují stejně
2. klikni na věc
3. vytěž jí za pár vteřin, dostaň peníze
4. za peníze kup "peta"
5. pet zvětší poškození a počet peněz za těžení
6. odemkni za peníze novou lokaci
7. vylepši svojí postavičku za gemy (peníze ale více vzácné)
8. přidej spoustu blbostí co stojí RBX
9. přidej GUI co ukazuje všechny ty "výhodné" věci za RBX
10. udělej to tak, aby se na to jednoduše kliknulo a koupilo
11. dítě na mobilu maminky klikne, koupí a bum
12. profit

Pet Simulator X - Hra která je jedna z nehranější na platformě vydělává 32 000 dollarů denně v robuxech(podle google search - podle mě to je celkem nepřesné) (32,000 United States Dollar se rovná 705,427.20 Czech Koruna)
<br>
Díky game developer programu v robloxu se dá veliký počet robuxů převest na realné peníze (https://en.help.roblox.com/hc/en-us/articles/115005718246-Developer-Exchange-Terms-of-Use)
<br>
Jestli se vám tohle nechce číst, 1 ROBUX(RBX) je $0.0035 (0.077 Kč) Hra denně vydělá kolem 3,75 - 15  milionů robuxů. Získat peníze za RBX jde jen jednou měsíčně, stačí aby vám bylo pouze 13 let, je tam určitý úrok pro transakci, nemůžou vyměnit peníze do všech zemí.
<br>
Samotná hra donedávna vydávala týdně update kde bylo vždy nové "speciální" vajíčko které se dalo koupit jen za RBX, a to většinou jedno vajíčko = 400RBX(zhruba 130Kč) ve kterém jsou malé šance na "Dobré pety" a obrovské šance na "průměrné pety". To v podstatě znamená že skoro každý týden vyhodili nové vajíčko které si lidi ve velkém množství kupovali aby mohli mít co nejdříve nového dobreho peta a prodat ho za co nejvíce gemů(jediná podstatná herní měna pro "trading")
